```python
import re
import pandas as pd
```

```python
f = open("WhatsApp Chat with BTech CSE 4th Year 2022-26.txt",'r',encoding = 'utf-8')
```

```python
data = f.read()
```

```python
print(data)
```

```python
pattern = r'\d{1,2}/\d{1,2}/\d{2},\s\d{1,2}:\d{2}\s(?:AM|PM)\s-\s'
```

```python
message = re.split(pattern, data)[1:]
message
```

```python
dates = re.findall(pattern,data)
dates
```

```python
df = pd.DataFrame({'user_message': message, 'message_date': dates})
#convert message data type
df['message_date'] = pd.to_datetime(df['message_date']
    .str.replace('\u202f', ' ', regex=False)  # replace narrow non-breaking space
    .str.replace('\u00a0', ' ', regex=False),  # replace normal non-breaking space                                
    format='%m/%d/%y, %H:%M %p - ')

df.rename(columns = {'message_date': 'date'}, inplace = True)
df.head(25)
```

```python
df.shape
```

# Seperate User and Messages

```python
users = []
messages = []
for message in df['user_message']:
    if not isinstance(message, str):
        message = str(message)
    entry = re.split(r'([\w\W]+?):\s',message)
    if entry[1:]:
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('group_notification')
        messages.append(entry[0])

df['users'] = users
df['messages'] = messages
df.drop(columns = ['user_message'], inplace = True)

df.head(10)
```

```python
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month_name()
df['day'] = df['date'].dt.day
df['hour'] = df['date'].dt.hour
df['minute'] = df['date'].dt.minute
df.head()
```

```python
df['hour_label'] = df['date'].dt.strftime('%I %p')
df.head(100)
```

```python
period = []
for hour in df['hour']:
    next_hour = (hour + 1) % 24
    
    # Format hours in 12-hour format with AM/PM
    start = f"{hour:02d}"
    end = f"{next_hour:02d}"
    
    start_label = pd.to_datetime(start, format="%H").strftime("%I %p")
    end_label = pd.to_datetime(end, format="%H").strftime("%I %p")

    period.append(f"{start_label} - {end_label}")

df['period'] = period
```

```python
df
```

```python
period = []
for hour in df['hour']:
    next_hour = (hour + 1) % 24
    
    # Convert start hour to 12-hour format
    start_hour_12 = hour % 12
    start_hour_12 = 12 if start_hour_12 == 0 else start_hour_12
    start_ampm = "AM" if hour < 12 else "PM"
    
    # Convert end hour to 12-hour format
    end_hour_12 = next_hour % 12
    end_hour_12 = 12 if end_hour_12 == 0 else end_hour_12
    end_ampm = "AM" if next_hour < 12 else "PM"
    
    period.append(f"{start_hour_12:02d} {start_ampm} - {end_hour_12:02d} {end_ampm}")

df['period'] = period
```

```python
from datetime import datetime
```

```python
period = []
for (hour_label,hour) in zip(df['hour_label'], df['hour']):
    if hour_label == '11 PM':
        period.append(str(hour_label) + "-" + str('00 AM'))
    elif hour_label == '11 AM':
        period.append(str(hour_label) + "-" + str('12 PM'))
    elif hour_label == '12 PM':
        period.append(str(hour_label) + "-" + str('1 PM'))
    elif hour_label == '00 AM':
        period.append(str(hour_label) + "-" + str('1 AM'))
    else:
        start = hour_label
        end = datetime.strptime(str((hour+1) % 24), "%H").strftime("%I %p")
        period.append(f"{start}-{end}")
```

```python
df['period'] = period
```
